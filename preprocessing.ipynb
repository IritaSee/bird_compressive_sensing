```python
import matplotlib.pyplot as plt
def plot_loss_and_accuracy(history):
  """Plots the training and validation loss and accuracy curves.

  Args:
    history: A Keras History object.
  """

  # Get the training and validation loss and accuracy.
  train_loss = history.history['loss']
  train_accuracy = history.history['acc']
  val_loss = history.history['val_loss']
  val_accuracy = history.history['val_acc']

  # Plot the loss and accuracy curves.
  plt.figure(figsize=(10, 6))

  # Plot the training loss.
  plt.subplot(2, 1, 1)
  plt.plot(train_loss, label='Training loss')
  plt.plot(val_loss, label='Validation loss')
  plt.title('Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.legend()

  # Plot the training accuracy.
  plt.subplot(2, 1, 2)
  plt.plot(train_accuracy, label='Training accuracy')
  plt.plot(val_accuracy, label='Validation accuracy')
  plt.title('Accuracy')
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.legend()

  # Show the plot.
  plt.tight_layout()
  plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

def plot_confusion_matrix(y_true, y_pred, classes, normalize=False):
  """Plots a confusion matrix.

  Args:
    y_true: The ground truth labels.
    y_pred: The predicted labels.
    classes: The labels of the different classes.
    normalize: Whether to normalize the confusion matrix.
  """

  cm = confusion_matrix(y_true, y_pred)

  if normalize:
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

  plt.figure(figsize=(10, 6))
  plt.imshow(cm, interpolation='nearest')
  plt.colorbar()
  plt.xticks(np.arange(len(classes)), classes)
  plt.yticks(np.arange(len(classes)), classes)
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
  plt.title('Confusion matrix')

  # Plot the text for each cell.
  for i in range(len(classes)):
    for j in range(len(classes)):
      text = cm[i, j]
      if normalize:
        text = f'{text:.2f}'
      else:
        text = f'{text:.0f}'
      plt.text(j, i, text, ha='center', va='center')

  # plt.tight_layout()
  # plt.show()

def calculate_precision_recall_f1_score(y_true, y_pred):
  """Calculates the precision, recall, and F1 score for multiclass classification.

  Args:
    y_true: The ground truth labels.
    y_pred: The predicted labels.

  Returns:
    precision: The precision score.
    recall: The recall score.
    f1_score: The F1 score.
  """

  rep = classification_report(y_true, y_pred, output_dict=True)
  precision = rep['macro avg']['precision']
  recall = rep['macro avg']['recall']
  f1_score = rep['macro avg']['f1-score']

  return precision, recall, f1_score
```

```python
#using waveform
#using specgram
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
tf.keras.backend.clear_session()


training_datagen = ImageDataGenerator(
    rescale = 1./255,
    # rotation_range=40,
    # width_shift_range=0.2,
    # height_shift_range=0.2,
    # shear_range=0.2,
    # zoom_range=0.2,
    # horizontal_flip=True,
    validation_split = 0.5)

train_generator = training_datagen.flow_from_directory('./waveform', batch_size=8,class_mode='categorical',target_size=(150,150))
valid_generator = training_datagen.flow_from_directory('./waveform', batch_size=2,class_mode='categorical',target_size=(150,150),subset='validation')
```

```python
# model = tf.keras.models.Sequential([
#         tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150,150,3)),
#         tf.keras.layers.MaxPooling2D(2,2),
#         tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
#         tf.keras.layers.MaxPooling2D(2,2),
#         tf.keras.layers.Flatten(),
#         tf.keras.layers.Dropout(0.3),
#         tf.keras.layers.Dense(512,activation='relu', kernel_regularizer='l1'),
#         tf.keras.layers.Dense(4,activation='softmax')
#     ])
model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(16, (3,3),  input_shape=(150,150,3)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(32, (3,3) ),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(16, (3,3), activation='relu', kernel_regularizer='l2'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(8, (3,3), activation='relu', kernel_regularizer='l1'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(128,activation='relu', kernel_regularizer='l2'),
        tf.keras.layers.Dense(4,activation='softmax')
    ])

model.compile(optimizer='adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['acc'])
hist = model.fit(train_generator,epochs=50, validation_data=valid_generator)
```

```python

# model4 = tf.keras.models.Sequential([
#         tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150,150,3)),
#         tf.keras.layers.MaxPooling2D(2,2),
#         tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
#         tf.keras.layers.MaxPooling2D(2,2),
#         tf.keras.layers.Flatten(),
#         tf.keras.layers.Dropout(0.3),
#         tf.keras.layers.Dense(512,activation='relu', kernel_regularizer='l1'),
#         tf.keras.layers.Dense(4,activation='softmax')
#     ])

model4 = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(16, (3,3),  input_shape=(150,150,3)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(32, (3,3) ),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(16, (3,3), activation='relu', kernel_regularizer='l2'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(8, (3,3), activation='relu', kernel_regularizer='l1'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(64,activation='relu'),
        tf.keras.layers.Dense(128,activation='relu', kernel_regularizer='l2'),
        tf.keras.layers.Dense(4,activation='softmax')
    ])

model4.compile(optimizer='adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['acc'])
hist = model4.fit(train_generator,epochs=50, validation_data=valid_generator)
```
